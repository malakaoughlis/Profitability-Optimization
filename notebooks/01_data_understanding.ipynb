## 7. Data Quality Checks

This section checks whether the dataset contains missing values, duplicate rows, or obvious quality issues.

```python
quality_checks = pd.DataFrame({
    "Check": [
        "Missing values",
        "Duplicated rows",
        "Duplicated Row.ID",
        "Negative sales",
        "Negative quantity",
        "Negative shipping cost",
        "Discount below 0",
        "Discount above 1"
    ],
    "Result": [
        df.isnull().sum().sum(),
        df.duplicated().sum(),
        df["Row.ID"].duplicated().sum(),
        (df["Sales"] < 0).sum(),
        (df["Quantity"] < 0).sum(),
        (df["Shipping.Cost"] < 0).sum(),
        (df["Discount"] < 0).sum(),
        (df["Discount"] > 1).sum()
    ]
})

quality_checks
```

### Observation

The dataset does not contain missing values or duplicated rows.  
No negative values were found in sales, quantity, shipping cost, or discount.

Negative profit values exist, but they are not data errors.  
They represent real business situations where some products were sold at a loss.

## 8. Date Coverage

This section checks the period covered by the dataset and verifies whether shipping dates are consistent with order dates.

```python
df_dates = df.copy()

df_dates["Order.Date"] = pd.to_datetime(df_dates["Order.Date"])
df_dates["Ship.Date"] = pd.to_datetime(df_dates["Ship.Date"])

date_summary = pd.DataFrame({
    "Metric": [
        "First order date",
        "Last order date",
        "First ship date",
        "Last ship date",
        "Orders with ship date before order date"
    ],
    "Value": [
        df_dates["Order.Date"].min(),
        df_dates["Order.Date"].max(),
        df_dates["Ship.Date"].min(),
        df_dates["Ship.Date"].max(),
        (df_dates["Ship.Date"] < df_dates["Order.Date"]).sum()
    ]
})

date_summary
```

```python
df_dates["Shipping.Delay"] = (df_dates["Ship.Date"] - df_dates["Order.Date"]).dt.days

df_dates["Shipping.Delay"].describe()
```

### Observation

The dataset covers orders from 2011 to 2014.  
Shipping dates may slightly extend beyond the last order date because some orders are shipped after being placed.

## 11. Potential Data Issues Identified

During the data understanding phase, a few points were identified for the cleaning phase:

1. The columns `Order.Date` and `Ship.Date` should be converted to datetime format.
2. The column `记录数` appears to be a record counter and does not provide analytical value.
3. Column names contain dots and inconsistent naming conventions, which should be cleaned.
4. A small number of rows have sales equal to zero and should be reviewed.
5. `Order.ID` is not unique because one order can contain multiple products.
6. `Row.ID` appears to be the unique identifier for each row.

```python
zero_sales_count = (df["Sales"] == 0).sum()

print(f"Number of rows with zero sales: {zero_sales_count}")

if zero_sales_count > 0:
    display(df[df["Sales"] == 0])
```

### Observation


Rows with zero sales should be reviewed if they exist.  
They should not be removed automatically without understanding whether they represent valid records, rounding issues, or data entry problems.
